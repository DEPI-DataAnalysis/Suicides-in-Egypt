```python
import pandas as pd
import numpy as np

df = pd.read_excel('suicides_in_egypt_ar.xlsx')
df
```

```python
from googletrans import Translator
import time

translator = Translator()

def safe_translate(text, src='ar', dest='en', max_retries=3, pause=0.3):
    if pd.isna(text):
        return text
    s = str(text)
    for i in range(max_retries):
        try:
            return translator.translate(s, src=src, dest=dest).text
        except Exception:
            time.sleep(pause * (i + 1))
    return s  

col_map = {col: safe_translate(col) for col in df.columns}
df.rename(columns=col_map, inplace=True)

for col in df.select_dtypes(include=['object']).columns:
    uniques = pd.Series(df[col].dropna().unique())
    print(f"Translating column '{col}' ({len(uniques)} unique values)...")
    cache = {}
    for val in uniques:
        cache[val] = safe_translate(val)
        time.sleep(0.05) 
    df[col] = df[col].apply(lambda x: cache.get(x, x))

output_path = 'suicides_in_egypt_en.xlsx'
df.to_excel(output_path, index=False)
print(f"Saved: {output_path}")
df.head()
```
